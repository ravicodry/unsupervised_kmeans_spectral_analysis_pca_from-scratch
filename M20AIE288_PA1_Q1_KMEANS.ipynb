```python
import random
init_center=random.sample(range(0,len(A)),2)
```

```python
#find the centroid of the selected center
centroid1=[]
for i in init_center:
    centroid1.append(df1.iloc[i,:])
```

```python
centroid1
```

```python
centroid1=np.array(centroid1)
```

```python
centroid1
```

```python
#function to caculate distance
def cal_distance(x1,x2):
    return (np.sum((x1-x2)**2))**0.5
```

```python
def nearest_Centroids(v, y):
    n_centroid = []
    for i in y:
        distance=[]
        for j in v:
            distance.append(cal_distance(i, j))
        n_centroid.append(np.argmin(distance))
        
    return n_centroid
```

```python
obtain_centroids = nearest_Centroids(centroid1, A)
obtain_centroids
```

```python
def mel_centroids(clusters, A):
    n_centroids = []
    n_df = pd.concat([pd.DataFrame(A), pd.DataFrame(clusters, columns=['cluster'])],
                      axis=1)
    for k in set(n_df['cluster']):
        current_cluster = n_df[n_df['cluster'] == k][n_df.columns[:-1]]
        cluster_mean = current_cluster.mean(axis=0)
        n_centroids.append(cluster_mean)
    return n_centroids
```

```python
import matplotlib.pyplot as plt
for i in range(10):
    obtain_centroids = nearest_Centroids(centroid1, A)
    centroid1 = mel_centroids(obtain_centroids, A)
    
plt.figure()
plt.scatter(np.array(centroid1)[:, 0], np.array(centroid1)[:, 1], color='red')
plt.scatter(A[:, 0], A[:, 1], alpha=0.1)
plt.show()
```

```python
plt.figure()
plt.scatter(np.array(centroid1)[:, 0], np.array(centroid1)[:, 1], color='black')
plt.scatter(A[:, 0], A[:, 1],c=obtain_centroids, cmap='cool' ,alpha=0.7)
plt.show()
```
